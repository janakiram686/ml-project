```python
df['Price Difference'] = df['SOLD PRICE'] - df['BASE PRICE']
print("\nAdded column for price difference:\n", df[['PLAYER NAME', 'BASE PRICE', 'SOLD PRICE', 'Price Difference']].head())
```

```python
df['Normalized Sold Price'] = (df['SOLD PRICE'] - df['SOLD PRICE'].min()) / (df['SOLD PRICE'].max() - df['SOLD PRICE'].min())
print("\nAdded column for normalized sold price:\n", df[['PLAYER NAME', 'SOLD PRICE', 'Normalized Sold Price']].head())
```
